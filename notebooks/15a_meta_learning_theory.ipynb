## 1. Meta-Learning in RL

**Goal**: Learn across tasks to quickly adapt to new tasks

**Standard RL**: Optimize for single task
$$\max_\theta \mathbb{E}_{\tau \sim \pi_\theta}[R(\tau)]$$

**Meta-RL**: Optimize for fast adaptation
$$\max_\theta \mathbb{E}_{\mathcal{T} \sim p(\mathcal{T})}[R(\mathcal{T}, \theta_{\mathcal{T}})]$$

where $\theta_{\mathcal{T}}$ = adapted parameters for task $\mathcal{T}$

**Applications:**
- Robotics (adapt to new objects)
- Games (learn new levels quickly)
- Personalization (adapt to users)

## 2. MAML (Model-Agnostic Meta-Learning)

**Key idea**: Learn initialization that adapts quickly with few gradient steps

**Algorithm:**
1. Sample batch of tasks $\{\mathcal{T}_i\}$
2. For each task $\mathcal{T}_i$:
   - Adapt: $\theta_i' = \theta - \alpha \nabla_\theta \mathcal{L}_{\mathcal{T}_i}(\theta)$
   - Evaluate on $\mathcal{T}_i$ with $\theta_i'$
3. Meta-update:
   $$\theta \leftarrow \theta - \beta \nabla_\theta \sum_i \mathcal{L}_{\mathcal{T}_i}(\theta_i')$$

**Intuition**: Find $\theta$ such that one gradient step leads to good performance

**Benefits:**
- Model-agnostic (works with any gradient-based learner)
- Few-shot adaptation
- State-of-the-art transfer

## 3. Context-Based Meta-RL

**RL² (Duan et al. 2016)**:
- RNN processes task experience
- Hidden state = task representation
- Policy: $\pi(a|s, h_t)$

**Benefits:**
- No explicit task ID needed
- Learns to explore and exploit
- End-to-end learning

## 4. Transfer Learning

**Types:**
1. **Fine-tuning**: Pre-train on source tasks, adapt to target
2. **Feature transfer**: Reuse representations
3. **Policy distillation**: Transfer via imitation

**Curriculum learning**:
- Order tasks from easy to hard
- Enables learning complex behaviors
- Teacher-student framework

## Summary

**Key concepts:**
1. Meta-learning = learning to learn
2. MAML optimizes for fast adaptation
3. Context-based methods (RL²)
4. Transfer learning strategies
