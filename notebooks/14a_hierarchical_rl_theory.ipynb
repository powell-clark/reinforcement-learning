## 1. Motivation for Hierarchical RL

**Challenges in flat RL:**
- Long horizons → sparse credit assignment
- Repeating primitives inefficient
- No skill reuse across tasks

**Hierarchical solution:**
- **High-level**: Choose which skill to execute
- **Low-level**: Execute skills (temporal abstractions)
- Enables: Planning, transfer, interpretability

## 2. Options Framework

**Option** $\omega = (\mathcal{I}, \pi, \beta)$:
- $\mathcal{I} \subseteq \mathcal{S}$: Initiation set
- $\pi: \mathcal{S} \times \mathcal{A} \to [0,1]$: Policy
- $\beta: \mathcal{S} \to [0,1]$: Termination condition

**SMDP (Semi-Markov Decision Process):**
$$Q(s, \omega) = \mathbb{E}[\sum_{t=0}^{\tau-1} \gamma^t r_t + \gamma^\tau Q(s_\tau, \omega') | s_0=s]$$

where $\tau$ = option duration

**Benefits:**
- Temporal abstraction
- Transfer across tasks
- Hierarchical planning

## 3. Feudal RL

**FeUdal Networks (FuN)** - Vezhnevets et al. 2017:

**Manager** (high-level):
- Sets goals $g_t$ for worker
- Operates at lower frequency
- Maximizes long-term reward

**Worker** (low-level):
- Follows goals from manager
- Operates at higher frequency
- Intrinsic reward: goal achievement

**Goal representation:**
$$g_t = f_{manager}(s_t)$$

**Worker intrinsic reward:**
$$r^{intrinsic}_t = \cos(\phi(s_{t+1}) - \phi(s_t), g_t)$$

where $\phi$ is learned embedding

## 4. Goal-Conditioned RL

**Universal Value Function Approximators (UVFA):**
$$Q(s, a, g)$$

Policy conditioned on goal:
$$\pi(a|s, g)$$

**Hindsight Experience Replay (HER):**
- Store failed trajectories
- Relabel with achieved goals
- Learn from "failures"

**Benefits:**
- Compositional tasks
- Zero-shot transfer
- Sample efficiency

**Key concepts:**
