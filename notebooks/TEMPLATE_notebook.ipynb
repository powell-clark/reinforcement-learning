### 4.3 Training Loop

````python
# Training hyperparameters
n_episodes = 1000
max_steps = 500

# Tracking
episode_returns = []
episode_lengths = []

# Training loop
for episode in range(n_episodes):
    obs, info = env.reset(seed=SEED + episode)
    episode_return = 0.0
    
    for step in range(max_steps):
        # Select action
        action = agent.select_action(obs)
        
        # Take step
        next_obs, reward, terminated, truncated, info = env.step(action)
        done = terminated or truncated
        
        # Update agent
        agent.update(obs, action, reward, next_obs, done)
        
        # Track
        episode_return += reward
        obs = next_obs
        
        if done:
            break
    
    # Record
    episode_returns.append(episode_return)
    episode_lengths.append(step + 1)
    
    # Log progress
    if (episode + 1) % 100 == 0:
        mean_return = np.mean(episode_returns[-100:])
        print(f"Episode {episode + 1}/{n_episodes}, Mean Return: {mean_return:.2f}")

env.close()
print("Training complete!")
````

## 5. Experiments

### 5.1 Visualize Training Progress

````python
# Plot training curve
plt.figure(figsize=(12, 5))

# Raw returns
plt.subplot(1, 2, 1)
plt.plot(episode_returns, alpha=0.3, label='Raw Returns')
# Moving average
window = 100
moving_avg = np.convolve(episode_returns, np.ones(window)/window, mode='valid')
plt.plot(range(window-1, len(episode_returns)), moving_avg, label=f'Moving Avg ({window})')
plt.xlabel('Episode')
plt.ylabel('Return')
plt.title('Training Progress')
plt.legend()
plt.grid(True, alpha=0.3)

# Episode lengths
plt.subplot(1, 2, 2)
plt.plot(episode_lengths, alpha=0.3)
plt.xlabel('Episode')
plt.ylabel('Episode Length')
plt.title('Episode Lengths')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

### 5.2 Evaluate Trained Agent

````python
# Evaluation
n_eval_episodes = 100
eval_returns = []

env = gym.make('CartPole-v1')

for episode in range(n_eval_episodes):
    obs, info = env.reset(seed=SEED + 1000 + episode)
    episode_return = 0.0
    done = False
    
    while not done:
        action = agent.select_action(obs)  # Deterministic evaluation
        obs, reward, terminated, truncated, info = env.step(action)
        episode_return += reward
        done = terminated or truncated
    
    eval_returns.append(episode_return)

env.close()

print(f"\nEvaluation Results ({n_eval_episodes} episodes):")
print(f"Mean Return: {np.mean(eval_returns):.2f} ± {np.std(eval_returns):.2f}")
print(f"Min Return: {np.min(eval_returns):.2f}")
print(f"Max Return: {np.max(eval_returns):.2f}")
````
